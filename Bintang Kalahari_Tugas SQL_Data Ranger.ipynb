```python
#Import Librabry
import sqlite3
import pandas as pd
import os
import shutil
```

SETUP

```python
if os.path.exists("data_penjualan.db"):
  os.remove("data_penjualan.db")
else :
  print("The File Does not Exist")

shutil.copyfile("data_penjualan_ori.db","data_penjualan.db")
```

TAMPILKAN DATA ORDER YANG MEMPUNYAI ORDER AMOUNT DIBAWAH 1000

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Buat Query SQL
result  = connection.cursor()
command = "SELECT * FROM orders WHERE ORD_AMOUNT<1000;"

# Eksekusi Query SQL
result.execute(command)

# Tampilkan Hasil dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```

TAMPILKAN INFORMASI CUSTOMER YANG MEMPUNYAI PAYMENT AMOUNT > 8000 DAN DIURUTKAN DARI A-Z

```python
#Buat Koneksi ke Database
connection = sqlite3.connect("data_penjualan.db")

#Buat Query SQL
result = connection.cursor()
command = "SELECT * FROM customer WHERE PAYMENT_AMT > 8000 ORDER BY CUST_NAME ASC;"

#Ekseskui Query SQL
result.execute(command)

#Tampilkan hasil dalam bentuk dataframe
results = result.fetchall();connection.close();
cols    = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results,columns =cols)

results_data
```
